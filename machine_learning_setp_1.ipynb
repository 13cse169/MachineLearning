```python
# Machine Learning is a subset of Artificial Intelligence where algorithms learn patterns from data to make decisions or predictions — without being given explicit rules.

# Traditional programming: Rules + Input Data → Output
# Machine Learning: Data + Output → Rules (learned automatically)
```

```python
# Types of Machine Learning:
#   The three main paradigms of learning from data

#   1. Supervised Learning — Learn from labeled examples
#     You provide input-output pairs. The model learns the mapping. Used when you know the "right answer" for training data.

#     Classification, Regression, Labeled Data

#   2. Unsupervised Learning — Find hidden patterns
#     No labels needed. The model discovers structure, clusters, and relationships in unlabeled data on its own.

#     Clustering, Dimensionality Reduction, Anomaly Detection

#   3. Reinforcement Learning — Learn by trial and error
#     An agent takes actions in an environment and learns from rewards/penalties. Used in games, robotics, autonomous systems.

#     Reward Signal, Agent & Environment, Policy Learning
```

```python
# Type :
#   Supervised
#   Unsupervised
#   Reinforcement
#   Semi-supervised

# Data :
#   Labeled (X, y)
#   Unlabeled (X only)
#   Rewards/Penalties
#   Partly labeled

# Goal :
#   Predict output
#   Find structure
#   Maximize reward
#   Leverage unlabeled

# Examples :
#   Spam filter, price prediction
#   Customer segments, anomalies
#   AlphaGo, self-driving cars
#   Medical imaging
```

```python
# The ML Workflow :
  # Every ML project follows these repeatable steps

# 1. Define the Problem
#   What are you predicting? Regression or classification? What data do you have?

# 2. Collect & Explore Data (EDA)
#   Understand distributions, find outliers, check correlations, visualize patterns.

# 3. Preprocess & Clean
#   Handle missing values, encode categoricals, scale features, remove outliers.

# 4. Feature Engineering
#   Create new features, select relevant ones, transform distributions.

# 5. Choose & Train Model
#   Select algorithm, train on training set, tune hyperparameters.

# 6. Evaluate
#   Measure performance on test set. Use appropriate metrics for the problem.

# 7. Deploy & Monitor
#   Serve predictions in production. Track model drift, retrain when needed.
```

```python
# 1. Supervised Learning
  # In this type, the model learns from labeled data (input + correct output).

# Example:
  # If you train a model with house features and their prices, it can predict the price of a new house.

# Common Algorithms:
  # Linear Regression
    # Used for predicting continuous values such as price, temperature, or salary.
    # Example: Predict house price based on size and location.

  # Logistic Regression
    # Used for classification problems.
    # Example: Email spam detection (spam / not spam).

  # Decision Tree
    # Creates a tree-like structure of decisions.
    # Example: Loan approval prediction.

  # Random Forest
    # Combines many decision trees to improve accuracy.

  # Support Vector Machine
    # Finds the best boundary that separates classes.

  # K-Nearest Neighbors
    # Classifies data based on similarity to nearby data points.

  # Naive Bayes
    # Based on probability theory and Bayes rule.
```

```python
# Lesson 1 — Linear Regression
  # Linear Regression predicts a continuous numeric value by finding a relationship between variables.

  # Example problems:
    # Predict house price
    # Predict salary from experience
    # Predict sales based on advertising budget

  # It tries to fit a straight line through the data.


  # The mathematical form is: y = mx + b
```

```python
# Goal: Predict a numeric value.

import pandas as pd
from sklearn.linear_model import LinearRegression

data = {
    "experience":[1,2,3,4,5],
    "salary":[30000,35000,50000,60000,70000]
}

df = pd.DataFrame(data)

x = df[['experience']]
y = df['salary']

model = LinearRegression()

model.fit(x, y)
# prediction = model.predict([[6]])

new_data = pd.DataFrame({'experience':[6]})
prediction = model.predict(new_data)

print("Predicted Salary: ", prediction)
print("Slope: ", model.coef_)
print("Intercept", model.intercept_)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import r2_score

data = fetch_california_housing()

data.keys()

df = pd.DataFrame(data.data, columns=data.feature_names)
df['Price'] = data.target

df.head()

x = df.drop("Price", axis=1)
y = df['Price']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

model = LinearRegression()
model.fit(x_train, y_train)

predictions = model.predict(x_test)
predictions[:5]

r2_score(y_test, predictions)
```

```python
# Lesson 2 — Logistic Regression
  # Unlike Linear Regression, which predicts numbers, Logistic Regression predicts categories.
  # Examples:
    # Spam detection: Spam / Not Spam, Disease prediction: Sick / Healthy, Loan approval: Approved / Rejected, Customer churn: Leave / Stay

  # Logistic Regression uses a sigmoid function to convert predictions into probabilities.
  # sigma(x)=\frac{1}{1+e^{-x}}
  # This function converts any value into a number between 0 and 1.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

data = {
    "hours":[1,2,3,4,5,6],
    "pass":[0,0,0,1,1,1]
}

df = pd.DataFrame(data)

X = df[['hours']]
y = df['pass']

model = LogisticRegression()

model.fit(X, y)

# model.predict([[3.5]])

new_data = pd.DataFrame({'hours':[3.5]})
model.predict(new_data)

y_pred = model.predict(X)
accuracy_score(y, y_pred)

# create range for smooth curve
hours_range = np.linspace(0,7,100).reshape(-1,1)

# probability of passing
prob = model.predict_proba(hours_range)[:,1]

plt.scatter(df['hours'], df['pass'], label="Fail")
plt.plot(hours_range, prob, label="Logistic Curve")

plt.title("Logistic Regression Graph")
plt.xlabel("Hours Studied")
plt.ylabel("Probability of Passing")

# decision boundary
plt.axhline(0.5, linestyle="--", label="Decision Boundary (0.5)")

plt.legend()
plt.show()
```

```python
# Important Concepts
  # Probability: Logistic regression predicts probability first.
  # Example: 0.73 → 73% chance of passing

# Decision Boundary
  # The boundary that separates classes.
  # Example:
  #   Pass if probability > 0.5
  #   Fail if probability < 0.5

# Model Evaluation
  # We use metrics like:
    # Accuracy: Percentage of correct predictions.
    # Confusion Matrix: Shows prediction performance.
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer

data = {
    "email":[
        "Win money now",
        "Hello how are you",
        "Claim your free prize",
        "Let's meet tomorrow",
        "Free lottery ticket"
    ],
    "spam":[1,0,1,0,1]
}

df = pd.DataFrame(data)

X = df['email']
y = df['spam']

# convert text to number
vectorizer = TfidfVectorizer()
X_vector = vectorizer.fit_transform(X)

# train model
model = LogisticRegression()
model.fit(X_vector, y)

# new email
new_email = pd.DataFrame({'email':["Free money offer"]})

new_email_vectorizer = vectorizer.transform(new_email)

model.predict(new_email_vectorizer)
```

```python
# Lesson 3 — K-Nearest Neighbors (KNN)
  # 1. What KNN Does:

    # KNN predicts the class of a new data point by looking at the nearest data points.
    # A new data point is classified based on the majority class of its closest neighbors.

    # Examples:
      # Movie recommendation: Find users with similar taste
      # Image classification: Compare similar images
      # Medical diagnosis: Compare patients with similar symptoms


  # 2. Simple Idea

    # Suppose we want to classify a fruit.

    # | Weight | Color  | Fruit  |
    # | ------ | ------ | ------ |
    # | 150g   | Red    | Apple  |
    # | 170g   | Red    | Apple  |
    # | 140g   | Yellow | Orange |
    # | 130g   | Yellow | Orange |

    # Now a new fruit appears:
      # Weight = 160g, Color = Red

    # KNN looks at the nearest fruits and decides the class.
    # If most neighbors are Apple, prediction = Apple.


  # 3. Distance Measurement

    # KNN finds nearest neighbors using distance.
    # Most common distance is Euclidean Distance.

    # d=\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}


  # 4. Choosing K

    # K means number of neighbors used.

    # Example:

      # K value	    Behavior
      # K = 1	      Very sensitive to noise
      # K = 3	      Balanced
      # K = 10	    More generalized

      # Usually choose odd numbers like: 3, 5, 7
```

```python
# 5. Python Implementation

import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

data = {
    "height":[150,160,170,180],
    "weight":[50,60,70,80],
    "gender":[0,0,1,1]
}

df = pd.DataFrame(data)

X = df[["height", "weight"]]
y = df["gender"]

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

# model.predict([[165, 65]])

new_person = pd.DataFrame({
    "height":[165],
    "weight":[65]
})

prediction = model.predict(new_person)
print("Predicted Gender: ", prediction[0])

# Example output: 0 → Female, 1 → Male
```

```python
# Lesson 4 — Decision Tree
  # 1. What Decision Tree Does

  # A Decision Tree makes decisions like a human using if–else conditions.

  # Example:

    # If income > 50k → Approve loan
    # Else → Reject loan

  # It builds a tree structure to reach a decision.

  # 2. How It Looks

  #         [Income > 50k?]
  #          /          \
  #       Yes            No
  #     Approve        Reject

  # Each step is called a node, and the final result is a leaf node.


  # 3. Example Problem: Predict whether a person will buy a product.

    # Age	Salary	Buy
    # 22	20000	  0
    # 35	50000	  1
    # 45	80000	  1
    # 25	30000	  0

    # The tree learns rules like:

    # If Salary > 40000 → Buy
    # Else → Not Buy


  # 4. Key Concepts
  #   Entropy

  #   Measures randomness in data.

  #   High entropy → mixed data
  #   Low entropy → pure data
  #   Information Gain

  #   Helps choose the best feature to split.

  #   Choose the feature that gives the most useful split
```

```python
# 5. Python Implementation: Using Scikit-learn.

import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = {
    "age": [22, 35, 45, 25],
    "salary": [20000, 50000, 80000, 30000],
    "buy": [0, 1, 1, 0]
}

df = pd.DataFrame(data)

X = df[["age", "salary"]]
y = df["buy"]

model = DecisionTreeClassifier()

model.fit(X, y)

new_data = pd.DataFrame([[30, 40000]], columns=["age", "salary"])

prediction = model.predict(new_data)

print("Prediction", prediction)
```

```python
# 6. Visualizing the Tree

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plot_tree(model, feature_names=["age","salary"], filled=True)
plt.show()
```

```python
# 7. Important Parameters

  # max_depth: Limits tree size.

  # DecisionTreeClassifier(max_depth=3)

  # criterion: How splits are chosen:

    # "gini" (default)
    # "entropy"

# 8. Advantages

  # ✔ Easy to understand
  # ✔ No need for feature scaling
  # ✔ Works for classification and regression

# 9. Limitations

# ✘ Can overfit easily
# ✘ Sensitive to small data changes
```

```python
# 10. Practice Problems: Problem 1

  # Predict loan approval

  # Income	Credit Score	Approved
  # 20000	  600	          0
  # 50000	  700	          1
  # 80000	  750	          1
  # 30000	  650	          0
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

data = {
    "income": [20000,50000,80000,30000],
    "credit_score": [600,700,750,650],
    "approved": [0,1,1,0]
}

df = pd.DataFrame(data)

X = df[["income", "credit_score"]]
y = df["approved"]

model = DecisionTreeClassifier()

model.fit(X, y)

new_data = pd.DataFrame([[80000, 650]], columns=["income", "credit_score"])

prediction = model.predict(new_data)

print("Prediction", prediction)

plot_tree(model, feature_names=["age","salary"], filled=True)
plt.show()
```
