```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Pie Chart :
  # Used to show percentage distribution.

labels = ["IT", "HR", "Finance", "Marketing"]
sizes = [40, 20, 25, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title("Department Distribution")

plt.show()
```

```python
# Box Plot:
  # Used to see data spread and outliers.

data = [10,12,15,18,20,22,25,30,35]

plt.boxplot(data)
plt.title("Data Distribution")

plt.show()
```

```python
# Area Chart:
  # Area charts show cumulative data trends.

x = [1,2,3,4,5]
y = [3,5,7,6,8]

plt.fill_between(x, y)
plt.title("Area Chart Example")

plt.show()
```

```python
# Chart Style Customization:
  # You can make graphs cleaner using built-in styles.

plt.style.use("ggplot")

x = [1,2,3,4]
y = [10,15,20,25]

plt.plot(x, y)
plt.title("Styled Plot")
plt.show()
```

```python
data = {
    "Year":[2020,2021,2022,2023],
    "Revenue":[50,65,80,95]
}

df = pd.DataFrame(data)

plt.plot(df["Year"], df["Revenue"], marker="o")

plt.title("Company Revenue Growth")
plt.xlabel("Year")
plt.ylabel("Revenue")

plt.show()
```
