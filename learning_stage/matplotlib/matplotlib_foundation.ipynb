```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
```

```python
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
# plot() -> create line graph
# title() -> set chart title
# xlabel() / ylabel() -> axis label
# show() -> display the graph
```

```python
x = [1, 2, 3, 4, 5]
y = [12, 34, 45, 60, 90]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.figure(figsize=(6,4))
plt.show()
```

```python
subject = ["Maths", "Science", "English", "History"]
marks = [85, 90, 75, 88]
```

```python
plt.bar(subject, marks)
plt.title("Student Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
# bar() -> creates vertical bars
# barh() -> creates horiontal bars
```

```python
plt.barh(subject, marks)
plt.show()
```

```python
emp = ["A", "B", "C", "D"]
salary = [30000, 45000, 25000, 40000]

plt.bar(emp, salary)
plt.title("Employees Salary")
plt.xlabel("Employees")
plt.ylabel("Salary")
plt.grid(axis="y")
plt.figure(figsize=(7,5))
plt.show()
```

```python
# Histogram (For Distribution)

data = np.random.randint(1, 100, 50)

plt.hist(data, bins=10)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# hist() creates distribution chart
# bins = number of intervals
# It groups values into ranges

# Very useful for: Salary distribution, Age distribution, Exam scores
```

```python
# Scatter Plot (Relationship Between 2 Variables)

x = [1,2,3,4,5,6]
y = [10,15,13,20,18,25]

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
# When to use: Height vs Weight, Experience vs Salary, Study hours vs Marks
# If points form a pattern → relationship exists.
```

```python
# Create a histogram of 100 random numbers.

data = np.random.randint(1, 500, 100)

plt.hist(data, bins=50)
plt.title("Histogram Chart")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
# Create a scatter plot for:

study_hours = [1,2,3,4,5,6,7]
marks = [40,50,65,70,75,85,90]

plt.scatter(study_hours, marks)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(axis="y")
plt.show()
```

```python
# Customization (Important)
```

```python
# Colors + Line Style + Markers

x = [1,2,3,4,5]
y = [10,20,15,25,30]

plt.plot(x, y, color="orange", linestyle="-.", marker="^")
plt.title("Customized Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
# Options you should know:

  # Colors: red, blue, green, black, orange

  # Linestyle: "-" solid, "--" dashed, "-." dash-dot, ":" dotted

  # Markers: "o" circle, "s" square, "^" triangle, "*" star, "x" cross
```

```python
#Adding Legend (Very Important)

x = [1,2,3,4,5]
y = [10,20,15,25,30]

plt.plot(x, y, label="Sales 2025")
plt.legend()
plt.show()
```

```python
# plt.legend(loc="upper right") // default
# plt.legend(loc="upper left")
# plt.legend(loc="lower center")
# plt.legend(loc="best") // auto picks least cluttered spot
```

```python
x = [1,2,3,4,5]
y = [10,20,15,25,30]
z = [12,18,20,28,38]

plt.plot(x, y, color="red", linestyle="--", marker="o", label="Product A")
plt.plot(x, z, color="green", linestyle="-.", marker="^", label="Product B")

plt.title("Product Line Chart")

plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.xlim(0, 6)
plt.ylim(0, 40)

plt.xticks([1,2,3,4,5])
plt.yticks([0,10,20,35,40])

plt.legend()
plt.grid(True)

plt.show()
```

```python
# Axis Limits: xlim, ylim
# Ticks Customization: xtricks, ytricks
```
