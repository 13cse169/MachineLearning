```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
```

```python
# Heatmap (Very Important)
  # A heatmap is used to visualize correlation between variables.
```

```python
correlation = df.corr(numeric_only=True)

sns.heatmap(correlation, annot=True)

plt.show()
```

```python
# What it shows:
  # correlation between numeric columns
  # values between -1 and 1
```

```python
# Meaning:
  # 1	strong positive relation
  # 0	no relation
  # -1	strong negative relation
```

```python
# Pairplot
  # Pairplot shows relationships between all numerical variables.
```

```python
sns.pairplot(df)
plt.show()
```

```python
# What you get:
  # scatter plots between each pair of variables
  # histogram on diagonals
# This is often the first visualization step in data analysis.
```

```python
# Distribution Plot
  # Used to see how data values are distributed.
```

```python
sns.histplot(df["total_bill"], kde=True)
plt.show()
```

```python
# This shows:
  # histogram
  # density curve

# You can quickly see:
  # skewness
  # spread of data
```

```python
# Regression Plot
  # Shows relationship + regression line.
```

```python
sns.regplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
# This helps understand:
  # whether variables have a linear relationship
  # direction of the relationship
# Very useful before building linear regression models.
```
