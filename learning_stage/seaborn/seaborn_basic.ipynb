```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Built-in Dataset
  # Seaborn provides some sample datasets.
```

```python
df = sns.load_dataset("tips")

df.head()
```

```python
# Scatter Plot:
  # Used to see the relationship between two variables.
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df)

plt.show()
```

```python
# Count Plot:
  # Used to count category frequency.
```

```python
sns.countplot(x="day", data=df)

plt.show()
```

```python
# Box Plot:
  # Used to view data distribution and outliers.
```

```python
sns.boxplot(x="day", y="total_bill", data=df)

plt.show()
```

```python
# Violin Plot:
  # Similar to box plot but shows distribution shape.
```

```python
sns.violinplot(x="day", y="total_bill", data=df)

plt.show()
```

```python
# Add Category Groups:
  # You can split data using hue.
```

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    data=df
)

plt.show()
```
