```python
# 1️⃣ Normal Distribution

# A normal distribution is the most common pattern in real-world data.

# It forms a bell-shaped curve.

# Example areas where it appears:
  # human height
  # exam scores
  # measurement errors

# Properties:
  # Mean = Median = Mode
  # Symmetrical distribution
  # Most values near the center

# In a normal distribution:
  # 68% of data lies within 1 standard deviation
  # 95% within 2 standard deviations
  # 99.7% within 3 standard deviations
```

```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(50, 10, 1000)

sns.histplot(data, kde=True)
plt.plot
```

```python
# 2️⃣ Skewness
  # Skewness tells us if data is tilted to one side.

# Types of skew
  # Positive skew (right skew)
  # Most values are small, with a few very large values.

# Example:
  # income distribution

# Negative skew (left skew)
# Most values are large, with a few very small values.

# Example:
  # easy exam scores

# Positive skew → tail on the right
# Negative skew → tail on the left
```

```python
data = [10, 12, 13, 14, 100]

df = pd.Series(data)

df.skew()
```

```python
# 3️⃣ Probability Basics
  # Probability measures how likely something is to happen.

# Formula:
  # Probability = Favorable Outcomes / total Outcomes

# Example:
  # Rolling a dice.

  # Possible outcomes:
    # 1 2 3 4 5 6

  # Probability of getting 3:
    # 1 / 6 = 0.166

# Probability values always lie between: 0 and 1
  # Where:
    # 0 → impossible event
    # 1 → certain event
```

```python
# 4️⃣ Central Limit Theorem (CLT)
  # This idea is very important in statistics and ML.
  # When you take many samples from any distribution, the mean of those samples will form a normal distribution.
  # Even if the original data is not normal.

# Example concept:
  # population distribution → random
  # sample means → normal distribution

# Why this matters:
  # helps estimate population parameters
  # used in hypothesis testing
  # used in many ML evaluation methods
```

```python
sample_means = []

for i in range(100):
  sample = np.random.randint(1, 100, 30)
  sample_means.append(np.mean(sample))

sns.histplot(sample_means, kde=True)
plt.show()
```
