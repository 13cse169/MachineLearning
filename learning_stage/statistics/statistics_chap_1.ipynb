```python
# Let’s begin Statistics for Data Science. This is the math used behind machine learning models and data analysis. You’ll mostly apply it while working with NumPy, Pandas, and visualization libraries like Seaborn.
```

```python
# 1. Mean (Average)
```

```python
# The mean is the average value of a dataset.

# Mean = Sum of values / Number of values

# Marks = [60, 70, 80, 90, 100]

# (60 + 70 + 80 + 90 + 100) / 5 = 80
```

```python
import numpy as np
from scipy import stats
```

```python
data = [60, 70, 80, 90, 100]

mean = np.mean(data)

mean
```

```python
# 2. Median
```

```python
# The median is the middle value when numbers are sorted.

# Data = [10, 20, 30, 40, 50]

# Median = 30

# Data = [10, 20, 30, 40]

# Median = (20 + 30) / 2 = 25
```

```python
data = [10, 20, 30, 40]
median = np.median(data)
median
```

```python
# 3. Mode
```

```python
# The mode is the value that appears most frequently.

# Data = [2,3,3,4,5,6]

# Mode = 3
```

```python
data = [2, 3, 3, 4, 5, 6]
mode = stats.mode(data)
mode
```

```python
# 4. Variance
```

```python
# Variance shows how spread out the data is.

# If variance is:
  # small → data points close together
  # large → data widely spread

# Data A: [48,50,52]
# Data B: [10,50,90]

# Both have same mean, but data B is much more spread out.
```

```python
data = [10,20,30,40,50]

variance = np.var(data)

variance
```

```python
# 5. Standard Deviation
```

```python
# Standard deviation shows average distance from the mean.

# Small SD → values close to mean
# Large SD → values far from mean
```

```python
data = [10,20,30,40,50]

std = np.std(data)

std
```

```python
# Mean → average value
# Median → middle value
# Mode → most frequent value
# Variance → spread of data
# Standard deviation → distance from mean
```
