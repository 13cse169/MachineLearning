```python
# Hypothesis Testing:
  # This concept is often used when analyzing data before training models with tools like Scikit-learn.
```

```python
# We’ll cover these ideas step by step:

  # What hypothesis testing is
  # Null hypothesis and alternative hypothesis
  # p-value
  # Confidence interval
  # z-test and t-test
```

```python
# 1️⃣ What is Hypothesis Testing?
  # Hypothesis testing is a statistical method used to check whether an assumption about data is likely true or not.

# Example situation:
  # A company claims: The average delivery time is 30 minutes.
  # You collect data and test whether this claim is correct.

# Hypothesis testing helps decide:
  # accept the claim
  # reject the claim
```

```python
# 2️⃣ Null Hypothesis (H₀) and Alternative Hypothesis (H₁)
  # In hypothesis testing, we always define two statements.

# Null Hypothesis (H₀)
  # The default assumption.
  # It usually means no effect or no difference.

# Example:
  # H₀: Average delivery time = 30 minutes

# Alternative Hypothesis (H₁)
  # This means something has changed or is different.

# Example:
  # H₁: Average delivery time ≠ 30 minutes

# Goal: Use data to decide whether to reject H₀.
```

```python
# 3️⃣ p-value
  # The p-value tells us how strong the evidence is against the null hypothesis.

# Typical rule used in data science:
  # p-value < 0.05 → Reject H₀
  # p-value ≥ 0.05 → Do not reject H₀

# Meaning:
  # small p-value → strong evidence against H₀
  # large p-value → not enough evidence to reject H₀

# Example:
  # p-value = 0.02

# Since 0.02 < 0.05, we reject the null hypothesis.
```

```python
# 4️⃣ Confidence Interval
  # A confidence interval gives a range where the true population value likely lies.

# Example:
  # Suppose the average height estimate is: 170 cm ± 2 cm

# Confidence interval: 168 cm – 172 cm

# If using 95% confidence level, it means:
# There is a 95% chance the real population mean lies in this range.
```

```python
# 5️⃣ Z-Test
  # A z-test is used when:
    # sample size is large (usually > 30)
    # population standard deviation is known

# Example:
  # Testing whether the average exam score differs from 70.

# Conceptually:
  # z = (sample_mean − population_mean) / standard_error
```

```python
# 6️⃣ T-Test
  # A t-test is used when:
    # sample size is small
    # population standard deviation is unknown

# Types of t-tests:
  # One-sample t-test
  # Independent t-test
  # Paired t-test
```

```python
from scipy import stats
```

```python
data = [65,70,72,68,74,69]

t_stat, p_value = stats.ttest_1samp(data, 70)

p_value
```

```python
# Null Hypothesis (H₀) → default assumption
# Alternative Hypothesis (H₁) → opposite claim
# p-value → evidence against H₀
# Confidence Interval → range for population value
# z-test → large samples
# t-test → small samples
```
